### Text Preprocessing


```python
import numpy as np
#Gensim 
import gensim.corpora 
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
#spacy
import spacy
#nltk
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#Downloading a list of stopwords in english
stop = stopwords.words('english')
```

Lemmatization takes into consideration the morphological analysis of the words. To do so, it is necessary to have detailed dictionaries which the algorithm can look through to link the form back to its lemma. 

Example: studies : study  , men : man

**en_core_web_sm** is a small English pipeline trained on written web text (blogs, news, comments), that includes vocabulary, syntax and entities.

```python
#Lemmatization using spacy

nlp = spacy.load("en_core_web_sm")
# data['Lemmatized'] = data['Abstract'].apply(lambda x: [item for item in x if item not in stop])
data['Lemmatized'] = data['Abstract'].apply(nlp)


```

```python
data
```

```python
# Tokenizing
# Lowercasing
# Removing punctuation
# by using simple_preprocess from gensim.utils

data['Lemmatized'] = data['Lemmatized'].apply(lambda x: simple_preprocess(str(x) , deacc=True))
```

```python
data
```

```python
# Removing Stopwords using nltk stopwords
data['Lemmatized'] = data['Lemmatized'].apply(lambda x: [item for item in x if item not in stop])
data['Lemmatized'] = data['Lemmatized'].apply(lambda x: [item for item in x if len(item) > 2])
```

```python
# data.to_csv('data.csv')
# data = pd.read_csv('data.csv')
data
```

Creating a bag of words dictionary that encapsulates the mapping between normalized words and their integer ids.

Bag of words is a representation of th text by assigning an ID to each word and the frequency of this word i the text.*italicized text*

```python
#Bag of words
id2word = gensim.corpora.Dictionary(data['Lemmatized'])
```

```python
bow_corpus = [id2word.doc2bow(doc) for doc in data['Lemmatized']]
```

### Creating an LDA Model using gensim

LDA is a simple but efficient topic model.

LDA is used to classify text in a document to a particular topic. It builds a topic per document model and words per topic model, modeled as Dirichlet distributions.

Each document is modeled as a multinomial distribution of topics and each topic is modeled as a multinomial distribution of words.

LDA assumes that every chunk of text we feed into it will contain words that are somehow related. Therefore choosing the right corpus of data is crucial.

```python
lda_model =  gensim.models.LdaMulticore(bow_corpus, 
                                   num_topics = 20, 
                                   id2word = id2word,                                    
                                   passes = 10,
                                   workers = 2,
                                   random_state = 0)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic: {idx} Words: {topic}")
    print("\n")
```

```python
# Compute Perplexity
print('\nPerplexity: ', lda_model.log_perplexity(bow_corpus))  # a measure of how good the model is. lower the better.

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=data['Lemmatized'], dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

#### We can conclude here that choosing 20 topics isn't optimum, and also we need todo bigram and trigram to collect thewords that come together 

```python
#Bigrams and trigrams
bigram = gensim.models.Phrases(data['Lemmatized'], min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data['Lemmatized']], threshold=100)

bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

data['TrigramBigram'] = trigram_mod[bigram_mod[data['Lemmatized']]]
```

```python
print(data['TrigramBigram'][0])
```

```python
#TF-IDF
from gensim.models import TfidfModel
id2word = gensim.corpora.Dictionary(data['TrigramBigram'])
bow_corpus = [id2word.doc2bow(doc) for doc in data['TrigramBigram']]

tfidf = TfidfModel(bow_corpus, id2word = id2word)
low_value = 0.03
words = []
words_missing_in_tfidf = []
for i in range(0,len(bow_corpus)):
  bow = bow_corpus[i]
  low_value_words = []
  tfidf_ids = [id for id, value in tfidf[bow]]
  bow_ids = [id for id , value in bow]
  low_value_words = [id for id, value in tfidf[bow] if value<low_value]
  drops = low_value_words + words_missing_in_tfidf
  for item in drops:
    words.append(id2word[item])
  words_missing_in_tfidf = [id for id in bow_ids if id not in tfidf_ids]
  new_bow = [b for b in bow if b[0] not in low_value_words and b[0] not in words_missing_in_tfidf]
  bow_corpus[i] = new_bow

```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
lda_model =  gensim.models.LdaMulticore(bow_corpus, 
                                   num_topics = 20, 
                                   id2word = id2word,                                    
                                   passes = 10,
                                   workers = 2,
                                   random_state = 0)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print("Topic: {} \nWords: {}".format(idx, topic ))
    print("\n")
```

### Model Evaluation

We will choose Perplexity and Coherence


#### Perplexity
Perplexity in NLP is the exponentiation of the entropy.

Less entropy (or less disordered system) is favorable over more entropy. Because predictable results are preferred over randomness. This is why people say **low perplexity is good and high perplexity is bad** since the perplexity is the exponentiation of the entropy.

It captures how surprised a model is of new data it has not seen before, and is measured as the normalized log-likelihood of a held-out test set.

However, recent studies have shown that predictive likelihood (or equivalently, perplexity) and human judgment are often not correlated, and even sometimes slightly anti-correlated.

#### Coherence
A set of statements or facts is said to be coherent, if they support each other.


C_v measure is based on a sliding window, one-set segmentation of the top words and an indirect confirmation measure that uses normalized pointwise mutual information and the cosine similarity.

#### I calculated coherence for multiple number of topics
